```python
# Q3.

# np.array()  -> creates a new numpy array from an object

x = [1, 2, 3]
arr = np.array(x)

# np.asarray() -> Converts the input to an array, but it does not create a new copy if the input is already an array.

x = np.array([1, 2, 3])
arr = np.asarray(x)

# np.asanyarray() -> is a NumPy function that converts an input to an array while preserving its subclass type if the input is already an array-like object.

x = np.matrix([[1, 2], [3, 4]])
arr = np.asanyarray(x)
```

```python
# shallow copy >> change in 1 array will lead to change in another
arr = np.array([1,2,3])
a = arr
a[0] = 100

print(a)
print(arr)
```

```python
# Deep copy >> The change in one array will not reflect in another array

arr = np.array([1,2,3])
a = arr.copy()
a[0] = 100

print(a)
print(arr)
```

```python
# Q14.
import seaborn as sns

np.random.seed(42)

x = np.random.rand(100) * 2 - 1
y = np.random.rand(100) * 2 - 1

df = pd.DataFrame({'x': x, 'y': y})

def get_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant 1'
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant 2'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

df['Quadrant'] = df.apply(get_quadrant, axis=1)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='deep')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')

dates = pd.date_range(start='2023-01-01', periods=30)
temperature = np.random.uniform(20, 35, size=30)
humidity = np.random.uniform(30, 70, size=30)

weather_df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

plt.subplot(1, 2, 2)
ax1 = plt.gca()
ax1.plot(weather_df['Date'], weather_df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(weather_df['Date'], weather_df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Add legends for both y-axes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Q15.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

mu, std = norm.fit(data)
x = np.linspace(min(data), max(data), 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

plt.show()
```

```python
# Q17.

np.random.seed(42)
x = np.random.rand(100) * 2 - 1
y = np.random.rand(100) * 2 - 1

df = pd.DataFrame({'x': x, 'y': y})

df['Quadrant'] = np.where((df['x'] >= 0) & (df['y'] >= 0), 'Quadrant 1',
                  np.where((df['x'] < 0) & (df['y'] >= 0), 'Quadrant 2',
                  np.where((df['x'] < 0) & (df['y'] < 0), 'Quadrant 3',
                  'Quadrant 4')))

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='deep')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')

plt.show()
```

```python
# Q18.

from math import pi
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 2 * pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='sin(X)')

p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

p.grid.grid_line_alpha = 0.5

show(p)
```

```python
# Q19.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap

output_notebook()

categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(1, 100, size=len(categories))

source = ColumnDataSource(data={'categories': categories, 'values': values})

color_mapper = linear_cmap('values', palette='Viridis256', low=min(values), high=max(values))

p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values', tools='')

p.vbar(x='categories', top='values', source=source, width=0.9,
        fill_color=color_mapper)

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

show(p)
```

```python
# Q20.

import numpy as np
import plotly.graph_objs as go
import plotly.io as pio

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = np.random.rand(100)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Line'))

fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

pio.show(fig)
```

```python
# Q21.

import numpy as np
import plotly.graph_objs as go
import plotly.io as pio

np.random.seed(42)
labels = [f'Category {i}' for i in range(1, 7)]
values = np.random.randint(1, 100, size=len(labels))

fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                               textinfo='percent+label',
                               hole=0.3)])

fig.update_layout(title='Interactive Pie Chart')

pio.show(fig)
```
